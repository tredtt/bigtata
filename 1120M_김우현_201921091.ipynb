데이터 시각화

```python
# Sorting co2 emission for 2019 in descending order
# 내림차순 정렬
co2_emission1 = co2_emission.sort_values(by = '2019', ascending = False)
co2_emission1['2019'].head()
```

```python
# sns 및 plt 속성 부여
sns.set_theme(style = 'darkgrid')
ax = plt.subplots(figsize=(10,35))
sns.set_color_codes('pastel')

#Plotting graph using seaborn barplot for co2 emission

sns.barplot(y = co2_emission.index,
            x = co2_emission['2019'])
plt.xlabel('CO2 Emission (metric ton per capita)')
plt.ylabel('Country')
plt.title("CO2 emission by countries in 2019")
```

```python
#연습 2018년도 co2 방출량 오름차순 정렬
# co2_emission2018 로 데이터프레임 지정

co2e2018 = co2_emission.sort_values(by = '2018', ascending =True)
co2e2018['2018'].head()
```

```python
# 연습
# sns 및 plt 속성 부여
#whitegrid 처리
sns.set_theme(style = 'whitegrid')

ax = plt.subplots(figsize=(10,35))
sns.set_color_codes('pastel')

#Plotting graph using seaborn barplot for co2 emission

sns.barplot(y = co2e2018.index,
            x = co2e2018['2018']) #2018년 입력
plt.xlabel('CO2 Emission (metric ton per capita)')
plt.ylabel('Country')
plt.title("CO2 emission by countries in 2018") #2018년 입력
```

(추가예제) CO2 배출량이 가장 높은 10개의 국가 년도별 방출량 시각화

```python
#Create a new column with the average CO2 emission from 1990-2019
# axis = 1 : 컬럼을 기준으로 평균값 분포 컬럼 만들기
co2_emission['avg'] = co2_emission.mean(axis = 1)

co2_emission
```

```python
#Select the 10 highest average CO2 emission
#avg 컬럼을 기준으로 내림차순 정렬

co2_avg = co2_emission.sort_values('avg', ascending = False).head(10)

co2_avg
```

```python
#시각화 자료의 간결성을 위해 avg 컬럼 삭제
co2_avg = co2_avg.drop('avg', axis = 'columns')

co2_avg
```

```python

#Plot time-series linegraph

fig, ax = plt.subplots(figsize =(20, 10))
n = 0
while n <10:
    sns.lineplot( x = co2_avg.columns, y = co2_avg.iloc[n], label = co2_avg.index[n])
    n = n + 1

plt.xlabel("Year")
plt.ylabel("CO2 Emission (metric ton per capita)")
plt.title("Top 10 Highest Average CO2 Emission per Capita Over Time")
```

```python
#연습 - CO2 방출이 가장 낮은 10개 국가 lineplot 출력
co2_avg = co2_emission.sort_values('avg', ascending = True).head(10)

co2_avg
```
